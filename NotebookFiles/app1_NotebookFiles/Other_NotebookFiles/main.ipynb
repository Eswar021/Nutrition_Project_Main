<h1>Building Recommendation system

```python
rec = dummy.copy()
rec.sample()
```

```python
rec.iloc[:,3:]
```

```python
rec['combined_text'] = rec['branded_food_category'].astype(str) + ' ' + rec['ingredients'].astype(str)
rec['combined_features'] = rec['branded_food_category'].astype(str) + ' ' + rec['ingredients'].astype(str) + ' ' + rec.iloc[:,3:].astype(str).agg(' '.join, axis=1)
```

```python
rec_features= rec[['ingredients','combined_features']].drop_duplicates()
rec_features.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(rec_features['combined_features'])
```

```python
X.shape
```

```python
rec_pca = PCA(500, svd_solver='arpack')
X = rec_pca.fit_transform(X)
```

```python
X.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(X, X)
```

```python
X
```

```python
cosine_sim
```

```python
sorted(list(enumerate(cosine_sim[0])),key=lambda x:x[1],reverse=True)
```

```python
rec_features = rec_features.reset_index()
```

```python
rec_features[rec_features.ingredients == "ORGANIC GREEN PEAS"].index[0]
```

```python
rec_features.iloc[1:6]
```

```python
rec_features
```

```python
def recommend(title):
    index = rec_features[rec_features.ingredients == title].index[0]
    distances=sorted(list(enumerate(cosine_sim[index])),key=lambda x:x[1],reverse=True)
    for i in distances[1:10]:
        print(rec_features.iloc[i[0]].ingredients)
```

```python
title = 'ORGANIC GREEN PEAS'
recommendations = recommend(title)

if recommendations is not None:
    print(recommendations)
```
