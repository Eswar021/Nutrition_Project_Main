```python
branded_food = pd.read_csv("branded_food.csv")
branded_food.head(2)
```

```python
branded_food.info()
```

```python
branded_food = branded_food[~branded_food.branded_food_category.isna()]
branded_food.info()
```

```python
branded_food.shape
```

```python
fruits = branded_food[branded_food.branded_food_category.str.lower().str.contains('fruit')]
fruits
```

```python
vegetables = branded_food[branded_food.branded_food_category.str.lower().str.contains('vegetable')]
vegetables
```

```python
df = pd.concat([fruits,vegetables])
df.head()
```

```python
df.branded_food_category.unique()
```

```python
fruits_and_vegetables_list = [
    'Canned Fruit',
    'Pre-Packaged Fruit & Vegetables',
    'Jam, Jelly & Fruit Spreads',
    'Fruit - Prepared/Processed (Shelf Stable)',
    'Fruits, Vegetables & Produce',
    'Fruit/Nuts/Seeds Combination',
    'Fruit – Prepared/Processed',
    'Fruit',
    'Frozen Fruit',
    'Wrapped Snacks - Fruit Bars',
    'Fresh Fruit and Vegetables',
    'Snack Foods - Dried Fruit',
    'Berries/Small Fruit',
    'Fruits Unprepared/Unprocessed (Shelf Stable)',
    'Fruit/Nuts/Seeds Mixes - Prepared/Processed (Shelf Stable)',
    'Fruit - Prepared/Processed (Frozen)',
    'Fruits Unprepared/Unprocessed (Frozen)']
FandV = [str.lower(x) for x in fruits_and_vegetables_list]
print(FandV)
```

```python
def ch(x):
    if x in FandV:
        return True
    else:return False
```

```python
fruits_vegs = branded_food[branded_food.branded_food_category.str.lower().apply(ch)]
fruits_vegs.head(2)
```

```python
fruits_vegs.info()
df = fruits_vegs[['fdc_id','ingredients', 'serving_size', 'branded_food_category']]
df
```

```python
cols = df.columns
cols
```

```python
FN = pd.read_csv('food_nutrient.csv')
FN.info()
FN.head(2)
```

```python
FN = FN[['id', 'fdc_id', 'nutrient_id', 'amount']]
FN.info()
```

```python
FN = FN.dropna()
FN.info()
```

```python
df2 = df.merge(FN,on = 'fdc_id', how = 'inner')
df2.info()
df2.head(2)
```

```python
df2.sort_values('amount',ascending=False)
```

```python
df3 = df2.drop_duplicates(subset=['fdc_id','ingredients', 'serving_size', 'branded_food_category'], keep = 'last').sort_values('amount',ascending=False)
df3.info()
df3.head(2)
```

```python
nutrient = pd.read_csv("nutrient.csv")
nutrient.info()
nutrient.head(2)
```

```python
df4 = df3.merge(nutrient,left_on='nutrient_id', right_on = 'id')
df4
```
