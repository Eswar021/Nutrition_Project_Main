<h2>Loading filtered data and Training the Model's<h2>

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import pickle
```

<h3>Training calories prediction model

```python
final = pd.read_csv("../../data/FilteredData/PredictionTable.csv")
```

```python
X = final.drop(['fdc_id','ingredients', 'Energy'],axis = 1)
Y = final.Energy.to_frame()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=30)
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
score = r2_score(y_test,pred)
print(score)
```

<h3>Creating similarity index and training recommendation model

```python
recommendation_df = pd.read_csv("../../data/FilteredData/RecommendationTable.csv")
dim_red_df = recommendation_df[['Calcium', 'Carbohydrate',
       'Cholesterol', 'Fatty_acids_total_saturated',
       'Fatty_acids_total_trans', 'Fiber', 'Iron',
       'Protein', 'Sodium', 'Sugars',
       'Fat', 'Energy']]
```

```python
pca = PCA(3)

dim_red = pca.fit_transform(dim_red_df)
similarity = cosine_similarity(dim_red)
sorted(enumerate(similarity[0]), key = lambda x: x[1], reverse= True)
```

```python
cls = recommendation_df[['Calcium', 'Carbohydrate',
       'Cholesterol', 'Fatty_acids_total_saturated',
       'Fatty_acids_total_trans', 'Fiber', 'Iron',
       'Protein', 'Sodium', 'Sugars',
       'Fat', 'Energy', 'ingredients']]
```

```python
LE = LabelEncoder()

x_cls = cls.drop('ingredients',axis = 1)
y_cls = cls.ingredients

x_cls = pca.transform(x_cls)
y_cls = LE.fit_transform(y_cls)
```

```python
#12 minutes to run
RF_clf = RandomForestClassifier(max_depth=15)
RF_clf.fit(x_cls, y_cls)
```
